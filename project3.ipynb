```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create dataframe from data set
path = 'Temperature_clean.csv'
original_df = pd.read_csv(path)
original_df.head()
```

```python
temp_mean = original_df.groupby(['State', 'Season'])[['Temp in F',  'Temp in C']].mean()
temp_mean.head()
```

```python
temp_mean.plot(kind="bar", figsize=(30,5))

# Set a title for the chart
plt.title("State Temperature Average")

plt.show()
```
