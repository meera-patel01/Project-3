```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/Temperature.csv')
```

```python
print('Temperature.csv') 
print(df)
```

```python
df.drop('Anomaly', inplace=True, axis=1)
```

```python
print(df)
```

```python
df.drop('Date', inplace=True,axis=1)
```

```python
'Value' == 'Temp in F'
```

```python
df.rename(columns = {'Value':'Temp in F'}, inplace = True)
df.dtypes
```

```python
df = df.dropna()
df = df.astype({"Month": 'int', "Year": 'int'})
```

```python
df
```

```python
df['Temp in F'] = pd.to_numeric(df['Temp in F'], errors='coerce')
```

```python
df['Temp in C'] = ((df['Temp in F'] - 32) * 5/9).round(1)
```

```python
df
```

```python
df = df[['State', 'Temp in F', 'Temp in C', 'Month','Year','Season']]
df
```

```python
df.to_csv('datasets/Temperature_clean.csv', index=False)
```
