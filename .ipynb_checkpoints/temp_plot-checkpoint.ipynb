```python
engine = create_engine("sqlite:///database/travel_db.sqlite")
```

```python
conn = engine.connect()
temperature = conn.execute(text("SELECT * FROM temperature_clean"))
for row in temperature:
    print(row)
```

```python
#gives warning but easier to get the dataframe from querying
#import so warning doesnt pop up
import warnings
warnings.filterwarnings('ignore')
conn = engine.raw_connection()
temp_f = pd.read_sql_query(sql="SELECT State, \"Temp in F\", Year, Season FROM temperature_clean WHERE State = \"Texas\"", con=conn)
temp_f = temp_f.groupby(['Season', 'Year'])[['Temp in F']].mean().round(0)
temp_f = temp_f.reset_index()
temp_f
```

```python
temp_c = pd.read_sql_query(sql=f"SELECT State, \"Temp in C\", Year, Season FROM temperature_clean WHERE State = \"Alaska\"", con=conn)
temp_c = temp_c.groupby(['Season', 'Year'])[['Temp in C']].mean().round(0)
temp_c = temp_c.reset_index()
temp_c.head()
```

```python
crime = pd.read_sql_query(sql=f"SELECT * FROM \"state-abbrevs\" INNER JOIN USCrime ON USCrime.State = \"state-abbrevs\".abbreviation WHERE \"state-abbrevs\".state = \"California\"",con=conn)
crime = crime.drop(['State', 'abbreviation', 'population', 'Total'], axis=1)
crime = crime.groupby(['year'])[['violent_crime', 'homicide', 'rape_revised', 'robbery', 'aggravated_assault',
                                'property_crime', 'burglary', 'larceny', 'motor_vehicle_theft']].mean().round(0)
crime = crime.reset_index()
crime
```

```python
column_names = list(crime.columns.values)
column_names.pop(0)
column_names
```

```python
column_values = crime.loc[crime.state == "Alabama", ['violent_crime', 'homicide', 'rape_revised', 'robbery', 'aggravated_assault',
                                'property_crime', 'burglary', 'larceny', 'motor_vehicle_theft']].values.flatten().tolist()
column_values
```

```python
location = pd.read_sql_query(sql=f"SELECT Location, State, Year, Address, Latitude, Longitude FROM location_clean WHERE State = \"California\"", con=conn)
def coord(c):
    return f"({c.Latitude}, {c.Longitude})"
location['Coordinate'] = location.apply(coord, axis=1)
location = location.drop_duplicates(subset=["Coordinate"])
location = location.drop(["Coordinate"], axis=1)
location
```
